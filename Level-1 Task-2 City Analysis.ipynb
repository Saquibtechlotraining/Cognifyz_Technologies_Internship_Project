‚Ä¢ Identify the city with the highest number of restaurants in the dataset.

‚Ä¢ Calculate the average rating for restaurants in each city.

‚Ä¢ Determine the city with the highest average rating.

###  Loading dataset:

```python
# Import all necessary libraries:

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset:

df = pd.read_csv(r"E:\Cognifyz_Internship\Dataset.csv")
df.head()
```

```python
# Number of rows and columns:

df.shape
```

```python
# Check the columns name and its datatype:

df.info()
```

## Identify the city with the highest number of restaurants.

```python
print(df['City'].value_counts()[[0]].index[0])
```

## Calculate the average rating for restaurants in each city.

```python
# Create a DataFrame that holds the average rating for restaurants in each city:

average_restaurant_rating = df.groupby('City').agg(Average_Rating=('Aggregate rating','mean')).sort_values('Average_Rating', ascending=True).reset_index()
average_restaurant_rating
```

```python
plt.subplots(figsize=(20,100))
average_restaurant_rating = df.groupby('City').agg(Average_Rating=('Aggregate rating','mean')).sort_values('Average_Rating', ascending=True).reset_index()
plt.barh(average_restaurant_rating['City'], average_restaurant_rating['Average_Rating'])
plt.title('Average rating for restaurants in each city')
plt.show()
```

## The city with the highest average rating.

```python
average_restaurant_rating = df.groupby('City').agg(Average_Rating=('Aggregate rating','mean')).sort_values('Average_Rating', ascending=False).reset_index()
print('The city with highest average rating:',average_restaurant_rating['City'][0])
```
