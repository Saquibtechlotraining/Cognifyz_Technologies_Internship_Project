‚Ä¢ Analyze the ratings and popularity of different restaurant chains.

### Loading Dataset

```python
# Import all necessary libraries:

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset:

df = pd.read_csv(r"E:\Cognifyz_Internship\Dataset.csv")
df.head()
```

```python
# Number of rows and columns:

df.shape
```

```python
# Check the columns name, its datatype and the Null values:

df.info()
```

## Identify if there are any restaurant chains present in the dataset.

```python
# Total counts of each restaurant:

restaurant_counts = df['Restaurant Name'].value_counts()
restaurant_counts
```

```python
# Identify restaurant chains

restaurant_chains = restaurant_counts[restaurant_counts > 1].index.tolist()
print('Restaurant chains:')
restaurant_chains
```

## Analyze the ratings and popularity of different restaurant chains.

```python
# Filter out the data for restaurant chain only:
restaurants_chain_data = df.loc[df['Restaurant Name'].isin(restaurant_chains)]

# Calculate average rating and popularity (i.e, Votes) for each restaurant chain
restaurants_chain_avg_rating = restaurants_chain_data.groupby('Restaurant Name')['Aggregate rating'].mean()
chain_total_votes = restaurants_chain_data.groupby('Restaurant Name')['Votes'].sum()

# Create a DataFrame and store the calculated data:
chain_information = pd.DataFrame({'Average rating' : restaurants_chain_avg_rating, 'Total votes' : chain_total_votes})
print('Restaurant chain ratings and Pouparity')
chain_information = chain_information.sort_values('Average rating', ascending=False) 
chain_information
```
